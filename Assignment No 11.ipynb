Q1. What is the concept of a metaclass?

- A metaclass in python is class that defines how class behaves. The class is itself instance of a metaclass.
- A Class is also an object, and just like any other object, it’s an instance of something called Metaclass. A special class type creates these Class objects. The type class is default metaclass which is responsible for making classes.
- The metaclass is responsible for the generation of classes, so we can write our custom metaclasses to modify the way classes are generated by performing extra actions or injecting code.

```python
def test_method(self):
    print("This is Test class method!")
 
# creating a base class
class Base:
    def myfun(self):
        print("This is inherited method!")
 
# Creating Test class dynamically using
# type() method directly
Test = type('Test', (Base, ), dict(x="kiran", my_method=test_method))
 
# Print type of Test
print("Type of Test class: ", type(Test))
 
# Creating instance of Test class
test_obj = Test()
print("Type of test_obj: ", type(test_obj))
 
# calling inherited method
test_obj.myfun()
 
# calling Test class method
test_obj.my_method()
 
# printing variable
print(test_obj.x)
```

What is the best way to declare a class's metaclass?

A way to declare a class’ metaclass is by using metaclass keyword in class definition.

```python
class meta(type):
    pass
class meta_class(metaclass=meta):
    pass
print(type(meta))
print(type(meta_class))
```
